```python
import matplotlib.pyplot as plt

import numpy as np

x = np.array([110, 70, 100, 80])

lab =  ["Python ", "Chemistry", "English", "Maths"]

col = ["blue", "orange", "Green", "red"]

abhi = [0.2, 0.0 , 0.0, 0.0]

plt.pie(x, labels = lab, colors = col, explode = abhi)

plt.show()
```
