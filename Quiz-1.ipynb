### Largest of three numbers

```python
num1=5
num2=3
num3=2
if num1>num2 and num1>num3:
    print(num1)
elif num2>num1 and num2>num3:
    print(num2)
elif num3>num1 and num3>num2:
    print(num3)
else:
    print('error')
```
