# conversion data types
## 1) int to float

```python
a=10
print(a)
print(type(a))
b=float(a)
print(b)
print(type(b))
```

## 2) float to int

```python
a=10.0
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
```

## 3) string to int/float

```python
a='2134'
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
b=float(a)
print(b)
print(type(b))
```

## 4) int to string

```python
a=1123
print(a)
print(type(a))
b=str(a)
print(b)
print(type(b))
```

## 5)Boolean

```python
a=bool(23)
print(a)
print(type(a))
b=int(a)
print(b)
print(type(b))
```

## 6)Boolean ex

```python
a=23
b=23
print(bool(a==b))
print(bool(a==23))
```
